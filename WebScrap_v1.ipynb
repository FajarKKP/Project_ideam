```python
import requests
import csv
from bs4 import BeautifulSoup

def get_all_sekolah_link(soup):
    #tabel data sekolah
    tabel = soup.find('table')

    url_list = []
    for tr in tabel.find_all('tr')[1:]:
        temp = tr.find_all('td')
        url = temp[2].a['href']
        nama_sekolah = temp[2].a.text
        url_list.append([url, nama_sekolah])
    return url_list

def get_kontak_sekolah(sekolah_url):
    sekolah_source = requests.get(sekolah_url).text
    sekolah_soup = BeautifulSoup(sekolah_source, 'lxml')

    kontak_div = sekolah_soup.find('div', id='tab_2')
    data_kontak = []
    
    try:
        for li in kontak_div.find_all('li'):
            info = li.font.text
            data_kontak.append(info)
        telp, fax, email = data_kontak
    except:
        telp, fax, email = '', '', ''
    return telp, fax, email

#halaman depan data semua sekolah
SMK_URL = 'http://datapokok.ditpsmk.net/sekolah?menu=0&ubah=&id_prov=310000&id_kab=&id_kec=&id_bidang=&id_program=&id_kompetensi=&sts=&cari='
source = requests.get(SMK_URL).text
soup = BeautifulSoup(source, 'lxml')

link_semua_smk = get_all_sekolah_link(soup)

with open('kontak.csv', 'w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['No.', 'Sekolah', 'No. Telp', 'No. Fax', 'Email'])
    counter = 0
    for smk in link_semua_smk:
        nama = smk[1]
        print(f'{counter + 1}. Mengakses data kontak {nama}')
        telp, fax, email = get_kontak_sekolah(smk[0])
        csv_writer.writerow([counter + 1, nama, telp, fax, email])
        counter += 1
    print(f'Total data kontak sekolah sebanyak {counter}')
```
